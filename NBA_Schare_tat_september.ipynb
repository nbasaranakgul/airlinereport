```python
# Analyze the distribution of ages for different occupations.
# Plot a dynamic bar chart with 'age' statistics grouped by 'occupation'.
# Here, we add a dropdown menu to select the value column to be plotted.

# graph_objects is used to create a figure and add traces to it (traces are the data points or lines that are plotted on the graph).
import plotly.graph_objects as go

fig = go.Figure()  # Create a figure

# set up a trace
# Add a bar trace to the figure, where the x-axis values are the index of the DataFrame, and the y-axis values are the values of the first column of the DataFrame.
fig.add_trace(
    go.Bar(
        x=occupation_stats.index,
        y=occupation_stats[occupation_stats.columns[0]],
        visible=True,
    )
)

# Here, we are adding a dropdown menu to the figure. The dropdown menu allows the user to select the value column to be plotted.
updatemenu = []
buttons = []

# button with one option for each value column in the DataFrame
for col in occupation_stats.columns:
    buttons.append(
        dict(
            method="restyle",  # The method parameter is set to 'restyle' to update the trace data.
            label=col,  # The label parameter is set to the column name.
            visible=True,  # The visible parameter is set to True to display the button.
            args=[  # The args parameter is a list of arguments to be passed to the method.
                {
                    "y": [
                        occupation_stats[col]
                    ],  # The y-axis values are set to the values of the selected column.
                    "x": [
                        occupation_stats.index
                    ],  # The x-axis values are set to the index of the DataFrame.
                    "type": "bar",  # The type is set to 'bar'.
                },
                [0],
            ],
        )
    )

# some adjustments to the updatemenus
# The updatemenu is a list of dictionaries that define the dropdown menu options.
updatemenu = []
your_menu = dict()
your_menu["buttons"] = buttons
your_menu["direction"] = "down"
your_menu["showactive"] = True

updatemenu.append(your_menu)  # Append the dropdown menu to the updatemenu list

# add dropdown menus to the figure
fig.update_layout(xaxis_title="Occupation", showlegend=False, updatemenus=updatemenu)
fig.write_html("plotly_plot-age_occupation.html")
fig.show()
```

# Machine Learning

## Missing values

```python
# Remind ourselves of the number of rows and columns in the dataframe
df.shape
```

```python
# Check the dataset information
# df.info provides information about the dataset, such as the number of non-null values and the data type of each column.
df.info()
```

* The number of non-null values in each column is equal to the number of rows in the dataset, which indicates that there are no 'null' values.
* But, there may be missing values represented by other text values such as 'NA', 'N/A', 'NaN', or '?'.
* For this dataset, the missing values are represented by '?'.

```python
# There is another way to check for missing values in the dataset.
# The isnull method to check for any missing values.
df.isnull().sum()
```

* Which still shows no missing values in the dataset because the missing values are represented by '?'.

```python
# Replace the '?' values in the dataset to 'null' or 'NaN' (Not a Number).
# np.nan is used to represent missing or null values in the dataset.
df[df == "?"] = np.nan
```

```python
# Check the dataset information after replacing '?' with 'NaN'
df.info()
```

* We can now see that number of non-null values in the 'workclass', 'occupation', and 'native.country' columns have decreased.

```python
# This shows the number of missing values in each column more clearly.
df.isnull().sum()
```

* There are missing values because of several reasons
    * Human errors
    * Technical issues
    * Privacy
    * Inherent nature
* Missing values are problematic because they can make it difficult to perform certain analyses.

### Imputation Methods

* Different strategies to take care of missing data. Two common strategies are:
    * Removing the rows altogether - will reduce your sample size
    * Impute the values - replace missing values with mean, median, or mode

```python
# Here, we impute the missing values with mode for the three columns
for col in ["workclass", "occupation", "native.country"]:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
# Now, check for missing values again
df.isnull().sum()
```

* There are no missing values in the dataset now.

```python
# In Machine Learning, we usually denote the input features with X and the output variable with y.

# X contains all the input features, that are all columns except the output column and the 'education.num' column (as per our preliminary analysis).
X = df.drop(["income", "education.num"], axis=1)

# y contains the output column, the income
y = df["income"]
```

```python
X.head()
```

```python
y.head()
```

```python
# To build and evaluate a machine learning model, we need to split the dataset into training and testing sets.
# This helps us evaluate the model's performance on unseen data.

# train_test_split function from the sklearn.model_selection module to split the dataset into training and testing sets
from sklearn.model_selection import train_test_split

# Here, we split the dataset into 80% training and 20% testing.
# The random_state parameter is used to set a seed for the random number generator used to split the dataset into training and testing sets.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Note that some of the input features are categorical.
# Since machine learning models work with numerical data, we need to encode these categorical features into numerical values.

# Use preprocessing.LabelEncoder from the sklearn module to encode the categorical columns
from sklearn import preprocessing

categorical = [
    "workclass",
    "education",
    "marital.status",
    "occupation",
    "relationship",
    "race",
    "sex",
    "native.country",
]
for feature in categorical:
    le = preprocessing.LabelEncoder()  # Create a LabelEncoder object
    X_train[feature] = le.fit_transform(
        X_train[feature]
    )  # Fit and transform the training data
    X_test[feature] = le.transform(X_test[feature])  # Transform the testing data
```

```python
# Let us visualize our training data after encoding the categorical features.
X_train.head()
```

* The categorical columns are now represented as numerical values.

```python
# Note that the numerical features are on different scales.
# Machine learning models work better when the features are on the same scale.

# We can scale the input features using StandardScaler from the sklearn.preprocessing module.
# StandardScaler standardizes the features by removing the mean and scaling to unit variance.
# Formula: z = (x - u) / s
# Here, z is the standard score, x is the value of the feature, u is the mean of the feature, and s is the standard deviation of the feature
# This ensures that the features are on the same scale.
from sklearn.preprocessing import StandardScaler

# Create a StandardScaler object
scaler = StandardScaler()

X_train = pd.DataFrame(
    scaler.fit_transform(X_train), columns=X.columns
)  # Fit and transform the training data
X_test = pd.DataFrame(
    scaler.transform(X_test), columns=X.columns
)  # Transform the testing data
```

```python
# View our training data after scaling the input features.
X_train.head()
```

* The input features are now standardized with a mean of 0 and a standard deviation of 1 (normally distributed).

```python
# We can now build a machine learning model using the training data.

# LogisticRegression from the sklearn.linear_model module to build a logistic regression model
from sklearn.linear_model import LogisticRegression

# accuracy_score from the sklearn.metrics module to evaluate the model
from sklearn.metrics import accuracy_score

logreg = LogisticRegression()  # Create a logistic regression model
logreg.fit(X_train, y_train)  # Fit the model on the training set
y_pred = logreg.predict(X_test)  # Make predictions on the testing set

print(
    "Logistic Regression accuracy score with all the features: {0:0.4f}".format(
        accuracy_score(y_test, y_pred)  # Calculate the accuracy score
    )
)
```

```python
# It is easy to try different models and compare their performance using sklearn.
# Let us try a Perceptron model.

# Perceptron is a linear classifier that makes predictions based on a linear combination of the input features.
from sklearn.linear_model import Perceptron

# Here, we build a perceptron model with a learning rate of 1 and a random state of 1.
# The learning rate is a hyperparameter that controls the step size during the training process.
ppn = Perceptron(eta0=1, random_state=1)

ppn.fit(X_train, y_train)  # Fit the model on the training set
y_pred = ppn.predict(X_test)  # Make predictions on the testing set

accuracy_score(y_pred, y_test)
```

```python
# We removed the 'education.num' column as it is highly correlated with the 'education' column.
# What about other columns that may not be useful for the model?

# PCA to help reduce the dimensionality of the dataset by projecting the input features onto the principal components.
# In simple terms, PCA helps us identify the most important features in the dataset.
from sklearn.decomposition import PCA

pca = PCA()
pca.fit_transform(X_train)  # Fit the PCA model on the training set

# The explained variance ratio of the principal components
# These values represent the percentage of the variance explained by each principal component
pca.explained_variance_ratio_
```

* We can see that the first principal component explains about 15.6% of the variance in the dataset while the last two principal components explains about 5.2% and 2.9% of the variance.

```python
X_train.columns
```

```python
# Let us remove the 'hours.per.week' and 'native.country' columns from input features and re-run the model.

# Running the model with 12 features
X = df.drop(["income", "hours.per.week", "native.country"], axis=1)
y = df["income"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

categorical = [
    "workclass",
    "education",
    "marital.status",
    "occupation",
    "relationship",
    "race",
    "sex",
]
for feature in categorical:
    le = preprocessing.LabelEncoder()
    X_train[feature] = le.fit_transform(X_train[feature])
    X_test[feature] = le.transform(X_test[feature])

X_train = pd.DataFrame(scaler.fit_transform(X_train), columns=X.columns)

X_test = pd.DataFrame(scaler.transform(X_test), columns=X.columns)

logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)

print(
    "Logistic Regression accuracy score with the first 12 features: {0:0.4f}".format(
        accuracy_score(y_test, y_pred)
    )
)
```

* The model accuracy improved from 0.8029 to 0.8247.
